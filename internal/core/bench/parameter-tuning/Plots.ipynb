```python
# Import the necessary modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('~/Desktop/bench_parameter_tuning_flat.csv', skiprows=9)
df = pd.DataFrame(data)
df.filter(like='Build', axis=0)

nlist = [1] # X
time = list(df["real_time"]) # Y0
time[:] = [x /1000.0/1000.0/1000.0 for x in time] # unit: second
  
# Plot the data using bar() method
x = np.arange(len(nlist))
plt.bar(x, time, width=0.35, label='time (s)', color='r', tick_label=nlist)

plt.title("FLAT Build")
# plt.xlabel("nlist")
plt.ylabel("time (s)")
  
# Show the plot
plt.legend()
plt.show()
```

```python
data = pd.read_csv('~/Desktop/bench_parameter_tuning_ivfflat.csv', skiprows=9)
df = pd.DataFrame(data)
df.filter(like='Build', axis=0)

nlist = list(df["nlist"]) # X
time = list(df["real_time"]) # Y0
time[:] = [x /1000.0/1000.0/1000.0 for x in time] # unit: second
size = list(df["Index Size"]) # Y1
size[:] = [x /1024.0/1024.0 for x in size] # unit: MiB
  
# Plot the data using bar() method
x = np.arange(len(nlist))
plt.bar(x, time, width=0.35, label='time (s)', color='r', tick_label=nlist)
plt.bar(x + 0.35, size, width=0.35, label='size (MiB)', color='g')

plt.title("IVF_FLAT Build")
plt.xlabel("nlist")
# plt.ylabel("Y NAME")
  
# Show the plot
plt.legend()
plt.show()
```

```python
data = pd.read_csv('~/Desktop/bench_parameter_tuning_ivfsq8.csv', skiprows=9)
df = pd.DataFrame(data)
df.filter(like='Build', axis=0)

nlist = list(df["nlist"]) # X0
nbits = list(df["nbits"]) # X1
nlist_nbits = nlist
nlist_nbits[:] = [str(x) + "," + str(nbits[index]) for index,x in enumerate(nlist_nbits)]


time = list(df["real_time"]) # Y0
time[:] = [x /1000.0/1000.0/1000.0 for x in time] # unit: second
size = list(df["Index Size"]) # Y1
size[:] = [x /1024.0/1024.0 for x in size] # unit: MiB
  
f = plt.figure()
f.set_figwidth(30)

# Plot the data using bar() method
x = np.arange(len(nlist))
plt.bar(x, time, width=0.35, label='time (s)', color='r', tick_label=nlist_nbits)
plt.bar(x + 0.35, size, width=0.35, label='size (MiB)', color='g')

plt.title("IVF_SQ8 Build")
plt.xlabel("nlist, nbits")
# plt.ylabel("Y NAME")

# Show the plot
plt.legend()
plt.show()
```

```python
data = pd.read_csv('~/Desktop/bench_parameter_tuning_ivfpq.csv', skiprows=9)
df = pd.DataFrame(data)
df.filter(like='Build', axis=0)

nlist = list(df["nlist"]) # X0
nbits = list(df["nbits"]) # X1
m = list(df["m"]) # X1
nlist_nbits_m = nlist
nlist_nbits_m[:] = [str(x) + "," + str(nbits[index]) for index,x in enumerate(nlist_nbits_m)]
nlist_nbits_m[:] = [str(x) + "," + str(m[index]) for index,x in enumerate(nlist_nbits_m)]


time = list(df["real_time"]) # Y0
time[:] = [x /1000.0/1000.0/1000.0 for x in time] # unit: second
size = list(df["Index Size"]) # Y1
size[:] = [x /1024.0/1024.0 for x in size] # unit: MiB
  
f = plt.figure()
f.set_figwidth(80)

# Plot the data using bar() method
x = np.arange(len(nlist))
plt.bar(x, time, width=0.35, label='time (s)', color='r', tick_label=nlist_nbits_m)
plt.bar(x + 0.35, size, width=0.35, label='size (MiB)', color='g')

plt.title("IVF_PQ Build")
plt.xlabel("nlist, nbits, m")
# plt.ylabel("Y NAME")

# Show the plot
plt.legend()
plt.show()
```
